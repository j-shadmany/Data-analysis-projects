```python
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None
```

```python
df = pd.read_excel("Online Retail.xlsx")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[df["Quantity"]>0]
df = df[df["UnitPrice"]>0]
df
```

```python
df.dropna(inplace= True)
df.info()
```

```python
df["Today"] = pd.to_datetime("today")
df.head(3)
```

```python
df["Recency"] = (df['Today'] - df['InvoiceDate']).dt.days
df.head(3)
```

```python
df["Total_price"] = df["Quantity"]* df["UnitPrice"]
df["CustomerID"] = df["CustomerID"].astype("int")
df.head(3)
```

```python
df["Frequency"] = df.groupby("CustomerID")["InvoiceNo"].transform("count")
df["Monetary_value"] = df.groupby("CustomerID")["Total_price"].transform("sum")
df.head()
```

```python
df = df.drop_duplicates(subset = ["CustomerID"], keep = "first")
df
```

```python
df = df.reset_index().set_index("CustomerID")
df.head()
```

```python
df2 = df[["Recency", "Frequency", "Monetary_value"]]
df2
```

```python
r_labels = range(4, 0, -1)
r_groups = pd.qcut(df2.Recency, q = 4, labels = r_labels)
f_labels = range(1, 5)
f_groups = pd.qcut(df2.Frequency, q = 4, labels = f_labels)
m_labels = range(1, 5)
m_groups = pd.qcut(df2.Monetary_value, q = 4, labels = m_labels)

df2['R'] = r_groups.values
df2['F'] = f_groups.values
df2['M'] = m_groups.values
df2
```

```python
df2['RFM_Score'] = df2[['R', 'F', 'M']].sum(axis = 1)
c_labels = ['Hibernating', 'About to sleep', 'Potential loyalist', 'Loyal', 'Champion']
df2['Customer_type'] = pd.qcut(df2.RFM_Score, q = 5, labels = c_labels)
df2.head()
```

```python
df2.to_csv("Final RFM.csv")
```

```python
df3 = df2[['R', 'F', 'M', 'RFM_Score', 'Customer_type']]
df3["R"] = pd.to_numeric(df3["R"], downcast="float")
df3["F"] = pd.to_numeric(df3["F"], downcast="float")
df3["M"] = pd.to_numeric(df3["M"], downcast="float")
df3['Customer_type'] = df3['Customer_type'].astype('string') 
df3.info()
df4 = df3.groupby('Customer_type').agg({"R" : "mean", "F" : "mean", "M" : "mean", 'RFM_Score': 'mean'})
df4
```

```python
df3.groupby('Customer_type').agg({"R" : "mean", "F" : "mean", "M" : "mean", 'RFM_Score': 'mean'}).plot(figsize=(8, 6), fontsize = 14)
plt.ylabel('Value', fontsize=18)
plt.xlabel('Customer type', fontsize=18)
plt.grid()
plt.title('Average score and rank for RFM and its parameters', fontsize=18)
```
